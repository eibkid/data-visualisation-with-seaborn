```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv("//users//uneku//downloads//1_titanic_modified_seaborn.csv")
titanic.head()
```

```python
titanic.info
```

```python
titanic = titanic.drop(['name', 'ticket', 'cabin'], axis=1)
```

```python
sns.catplot('sex', data=titanic, kind='count')
```

```python
sns.catplot('class', data=titanic, hue='sex', kind='count')
```

```python
sns.catplot('class', data=titanic, hue='embark_town', kind="count")
```

```python
sns.catplot('class', data=titanic, hue='alive', kind="count")
```

```python
sns.catplot('class', data=titanic, kind="count")
```

```python
sns.catplot('class', data=titanic, hue='survived', kind="count")
```

```python
sns.distplot(titanic.age.dropna())
plt.show()
```

```python
distribution = sns.FacetGrid(titanic, row='survived', col='class')
distribution.map(sns.distplot, "age")
plt.show()
```

```python
sns.heatmap(titanic.corr(), annot=True, fmt=".2f")
plt.show()
```
